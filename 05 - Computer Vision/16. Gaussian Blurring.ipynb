# Mặt nạ Gaussian (Bộ lọc Gaussian)

![title](img/gauss.png)

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
path = '.\\img\\sunflower01.jpg'
im = Image.open(path).convert('L')
arr = np.array(im)
```

## Thay đổi sigma

```python
blurred_im1 = ndimage.gaussian_filter(arr, sigma=1, mode='constant')
blurred_im2 = ndimage.gaussian_filter(arr, sigma=3, mode='constant')
```

```python
fig, (sub1, sub2, sub3) = plt.subplots(1, 3, figsize=(10,8))

sub1.imshow(im, cmap='gray')
sub2.imshow(blurred_im1, cmap='gray')
sub3.imshow(blurred_im2, cmap='gray')


plt.show()
```

## Thay đổi mode

```python
blurred_im3 = ndimage.gaussian_filter(arr, sigma=1, mode='reflect')
blurred_im4 = ndimage.gaussian_filter(arr, sigma=1, mode='nearest')
blurred_im5 = ndimage.gaussian_filter(arr, sigma=1, mode='mirror')
blurred_im6 = ndimage.gaussian_filter(arr, sigma=1, mode='wrap')
```

```python
fig, subs = plt.subplots(2, 2, figsize=(8,6))

sub1, sub2, sub3, sub4 = subs.flatten()

sub1.imshow(blurred_im3, cmap='gray')
sub2.imshow(blurred_im4, cmap='gray')
sub3.imshow(blurred_im5, cmap='gray')
sub4.imshow(blurred_im6, cmap='gray')

plt.show()
```
