# Các bộ lọc đạo hàm

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
path = '.\\img\\object01.jpg'
im = Image.open(path).convert('L')
arr = np.array(im)
```

## Bộ lọc Sobel

### Xem chi tiết tại: https://www.youtube.com/watch?v=Yz7h9L4gecQ

![title](img/sobel.png)

```python
sobel_gradient_x = ndimage.sobel(im, axis=0)
sobel_gradient_y = ndimage.sobel(im, axis=1)

sobel_gradient_magnitude = np.sqrt(sobel_gradient_x**2 + sobel_gradient_y**2)
```

```python
fig1, subs = plt.subplots(1, 4, figsize=(12,12))

sub1, sub2, sub3, sub4 = subs
sub1.imshow(arr, cmap='gray'); sub1.set_title('ORIGIN')
sub2.imshow(sobel_gradient_x, cmap='gray'); sub2.set_title('GRADIENT X')
sub3.imshow(sobel_gradient_y, cmap='gray'); sub3.set_title('GRADIENT Y')
sub4.imshow(sobel_gradient_magnitude, cmap='gray'); sub4.set_title('GRADIENT MAGNITUDE')

for sub in subs:
    sub.axis('off')
```

## Bộ lọc Prewitt

### Xem chi tiết tại: https://www.youtube.com/watch?v=fREYtbQ405E

![title](img/prewitt.png)

```python
# Tính gradient theo hướng x (đạo hàm riêng theo x) bằng mặt nạ Prewitt x
prewitt_gradient_x = ndimage.convolve(arr, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tính gradient theo hướng y (đạo hàm riêng theo y) bằng mặt nạ Prewitt y
prewitt_gradient_y = ndimage.convolve(arr, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
```

```python
# HOẶC
prewitt_gradient_x = ndimage.prewitt(arr, axis=0)
prewitt_gradient_y = ndimage.prewitt(arr, axis=1)

prewitt_gradient_magnitude = np.sqrt(prewitt_gradient_x**2 + prewitt_gradient_y**2)
```

```python
fig2, subs = plt.subplots(1, 4, figsize=(12,12))

sub1, sub2, sub3, sub4 = subs
sub1.imshow(arr, cmap='gray'); sub1.set_title('ORIGIN')
sub2.imshow(prewitt_gradient_x, cmap='gray'); sub2.set_title('GRADIENT X')
sub3.imshow(prewitt_gradient_y, cmap='gray'); sub3.set_title('GRADIENT Y')
sub4.imshow(prewitt_gradient_magnitude, cmap='gray'); sub4.set_title('GRADIENT MAGNITUDE')

for sub in subs:
    sub.axis('off')
```
