```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '.\\img\\face01.jpg'
im = Image.open(path)
gray_im = im.convert('L')
```

```python
im_arr = np.array(im)
im_arr.shape
```

```python
gray_arr = np.array(gray_im)
gray_arr.shape
```

# BIẾN ĐỔI ẢNH XÁM

```python
inverted_arr = 255 - gray_arr  # 255 - giá trị pixel
squared_arr = gray_arr**2
clamped_arr = np.clip(gray_arr, 100, 200)  # giá trị pixel >= 100 và <= 200
```

```python
fig, subs = plt.subplots(nrows=2, ncols=2, figsize=(8,4))
fig.subplots_adjust(wspace=0.1, hspace=0.5)

# subs là 1 mảng 2x2 (nrows x ncols)
sub1, sub2, sub3, sub4 = subs.flatten()

sub1.imshow(gray_arr, cmap='gray')
sub2.imshow(inverted_arr, cmap='gray')
sub3.imshow(squared_arr, cmap='gray')
sub4.imshow(clamped_arr, cmap='gray')

sub1.set_title('ORIGINAL IMAGE')
sub2.set_title('INVERTED IMAGE')
sub3.set_title('SQUARED IMAGE')
sub4.set_title('CLAMPED IMAGE')
```
