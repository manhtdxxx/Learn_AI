```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

# LƯỢC ĐỒ ẢNH

# LƯỢC ĐỒ HISTOGRAM

```python
path = '.\\img\\sunflower02.jpg'
img = Image.open(path)
```

```python
plt.imshow(img)
```

```python
gray_img = img.convert('L')
plt.imshow(gray_img, cmap='gray')
```

```python
img_arr = np.array(img)
img_arr.shape, img_arr[0].shape, img_arr[0]  
# Image contains 333 pixels in height, 500 pixels in width, and each pixel having 3 color channels (RGB)
```

```python
gray_img_arr = np.array(gray_img)
gray_img_arr.shape, gray_img_arr
```

```python
plt.hist(gray_img_arr.flatten(), bins=128)
plt.title('BIỂU ĐỒ HISTOGRAM CỦA ẢNH')
plt.xlabel('GIÁ TRỊ CỦA PIXEL')
plt.ylabel('SỐ LƯỢNG')
```

# ĐƯỜNG BIÊN CỦA ẢNH

```python
path1 = '.\\img\\face01.jpg'
im = Image.open(path1).convert('L')

plt.imshow(im, cmap='gray')

# thêm hình nữa
plt.figure()  # tránh đè lên cái trước
plt.gray()  # đặt ảnh màu xám
plt.axis('equal')  # cho 2 trục bằng nhau
plt.contour(im, origin='image')
```
