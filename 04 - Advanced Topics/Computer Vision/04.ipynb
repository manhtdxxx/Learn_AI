# CẮT ẢNH = CROP()

```python
width, height = img.size

left = 20  # min = 0
top = 20  # min = 0
right = 200  # max = 320
bottom = 160  # max = 240

# tạo ảnh mới
cropped_img = img.crop((left, top, right, bottom))
cropped_img
```

```python
cropped_img.size
```

# DÁN ẢNH VÀO ẢNH = PASTE()

```python
left = 20
top = 160
img.paste(cropped_img, (left, top))  # dán vào ảnh gốc
img
```
