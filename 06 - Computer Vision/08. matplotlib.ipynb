```python
import matplotlib.pyplot as plt
```

# LINE CHART

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 16, 20, 30, 25]
y2 = [20, 15, 8, 9, 20]

plt.plot(x, y1, 'bo-', label = 'Dữ Liệu')
plt.plot(x, y2, 'rs-', label = 'Dữ Liệu')
plt.legend()  # hiện label

plt.xlabel('Trục X')
plt.ylabel('Trục Y')
```

# BAR CHART

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color='g', alpha=0.3)  # alpha là độ đậm nhạt
plt.xlabel('CATEGORIES')
plt.ylabel('VALUES')
```

# PIE CHART

```python
my_labels = ['A', 'B', 'C', 'D']
my_sizes = [15, 30, 45, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(my_sizes, labels=my_labels, colors=my_colors, autopct='%1.2f%%')
plt.legend()
plt.title('PIE CHART')
```
