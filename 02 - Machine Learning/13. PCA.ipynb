```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
digit = load_digits()
dir(digit)
```

```python
df = pd.DataFrame(digit.data,columns=digit.feature_names)
df['target'] = digit.target
df['target_names'] = df.target.apply(lambda x:digit.target_names[x])
df
```

```python
df.describe()
```

```python
X = digit.data

scaler = StandardScaler().set_output(transform='pandas')

X_scaled = scaler.fit_transform(X)
X_scaled
```

# Training without PCA

```python
Y = digit.target

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, train_size=0.8, random_state=0)
X_train.shape, X_test.shape
```

```python
np.unique(Y_train, return_counts=True)
```

```python
lor = LogisticRegression()
lor.fit(X_train, Y_train)

lor.score(X_train, Y_train), lor.score(X_test, Y_test)
```

# Training using PCA

*1. Scale features before applying PCA*

*2. Score might drop*

```python
pca = PCA(n_components=16).set_output(transform='pandas')

X_reduced = pca.fit_transform(X_scaled)
X_reduced
```

```python
pca.n_features_
```

```python
pca.explained_variance_.shape, pca.explained_variance_
```

```python
variance_ratio = pca.explained_variance_ratio_
variance_ratio
```

```python
np.sum(variance_ratio)
```

```python
Y = digit.target

X_train, X_test, Y_train, Y_test = train_test_split(X_reduced, Y, train_size=0.6,random_state=0)
X_train.shape, X_test.shape
```

```python
np.unique(Y_train, return_counts=True)
```

```python
lor = LogisticRegression()
lor.fit(X_train, Y_train)

lor.score(X_train, Y_train), lor.score(X_test, Y_test)
```
