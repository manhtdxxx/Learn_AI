```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# Data

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

# From Scratch

```python
# y = ax + b
def gradient_descent(x, y, iterations, learning_rate):
    a = b = 0
    n = len(x)

    for i in range(1, iterations+1):
        y_predicted = a*x + b
        
        mse = (1/n)*sum((y-y_predicted)**2)
        
        plt.plot(x, y, 'r*')
        plt.plot(x, y_predicted, 'b-.')

        if i % 100 == 0:
            print('iterations = {:<5}, a = {:<20}, b = {:<20}, MSE = {:<20}'.format(i, a, b, mse))
        
        a_derivative = (-2/n)*sum(x*(y - y_predicted))
        b_derivative = (-2/n)*sum(y - y_predicted)

        if i == iterations:  # để không in th a, b nè
            break

        a = a - (learning_rate * a_derivative)
        b = b - (learning_rate * b_derivative)

    return a, b, mse
```

```python
gradient_descent(x, y, 1000, 0.01)
```
