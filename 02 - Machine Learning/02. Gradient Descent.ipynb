```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# y = ax + b
def gradient_descent(x,y,iterations,learning_rate):
    a = b = 0
    n = len(x)
    for i in range(iterations):
        y_predicted = a*x + b
        
        mse = (1/n)*sum((y-y_predicted)**2)
        
        plt.plot(x,y,'r*')
        plt.plot(x,y_predicted,'b-.')
        print('a={}, b={}, iterations={}, cost={}'.format(a,b,i,mse))
        
        a_derivative = (-2/n)*sum(x*(y - y_predicted)) # khi a thay đổi 1 đv thì mse thay đổi a_derivate đv
        b_derivative = (-2/n)*sum(y - y_predicted)
        a = a - (learning_rate * a_derivative) # giảm a để giảm mse
        b = b - (learning_rate * b_derivative)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y,10000,0.01)
```

### Using sklearn

```python
x = np.array([[1,2,3,4,5]]).T
y = np.array([5,7,9,11,13])
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.score(x,y)
```

```python
lr.intercept_
```

```python
y_p = lr.predict(x)
mean_squared_error(y,y_p)
```
