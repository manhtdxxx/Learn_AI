# Difference between KFold and StratifiedKFold

```python
# KFold không nên dùng do hoàn toàn lấy theo thứ tự, và không lấy class theo tỉ lệ, 
# có thể trong 1 lần lấy, thì lại toàn class 0 hoặc 1 thì sao???
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)
print(k_fold)
print()

for train_index, test_index in k_fold.split(digit.data, digit.target):
    print(train_index.shape, test_index.shape)
    
    Y_train = digit.target[train_index]
    Y_test = digit.target[test_index]

    print(np.unique(Y_train, return_counts=True))
    print(np.unique(Y_test, return_counts=True))
  
    print('~'*100)

# ĐỂ Ý COUNTS CỦA TỪNG CLASS CỦA Y_TRAIN KÌA TRONG CẢ 5 LẦN KÌA!!!
```

```python
# trong 1 lấy lấy thì lấy các class chia theo tỉ lệ với số lần xuất hiện của từng class trong dataset 
stratified_k_fold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
print(stratified_k_fold)
print()

for train_index, test_index in stratified_k_fold.split(digit.data, digit.target):
    print(train_index.shape, test_index.shape)
    
    Y_train = digit.target[train_index]
    Y_test = digit.target[test_index]

    print(np.unique(Y_train, return_counts=True))
    print(np.unique(Y_test, return_counts=True))
  
    print('~'*100)
    
# ĐỂ Ý COUNTS CỦA TỪNG CLASS CỦA Y_TRAIN KÌA TRONG CẢ 5 LẦN KÌA!!!
```

# Now, It's time for coding

```python
def model_score(model, X_train, Y_train, X_test, Y_test):
    model.fit(X_train, Y_train)
    return model.score(X_test, Y_test)
```

```python
lor = LogisticRegression()
svc = SVC()
rfc = RandomForestClassifier(n_estimators=33)
```

```python
scores_lor = []
scores_svc = []
scores_rfc = []
```

```python
skf = StratifiedKFold(n_splits=5)

for train_index, test_index in skf.split(digit.data, digit.target):
    X_train, X_test, Y_train, Y_test = digit.data[train_index], digit.data[test_index], \
                                    digit.target[train_index], digit.target[test_index]
    
    scores_lor.append(model_score(lor, X_train, Y_train, X_test, Y_test))  
    scores_svc.append(model_score(svc, X_train, Y_train, X_test, Y_test))  
    scores_rfc.append(model_score(rfc, X_train, Y_train, X_test, Y_test))  
```

```python
print(scores_lor)
print(scores_svc)
print(scores_rfc)
```

```python
from statistics import mean
print(mean(scores_lor))
print(mean(scores_svc))
print(mean(scores_rfc))
```

### Using sklearn.model_selection.cross_val_score

```python
X = digit.data
Y = digit.target
```

```python
print(cross_val_score(lor, X, Y, cv=5))  # return an array of scores of the testing set
print(cross_val_score(svc, X, Y, cv=5))
print(cross_val_score(rfc, X, Y, cv=5))
```

```python
print(cross_val_score(lor, X, Y, cv=5).mean())
print(cross_val_score(svc, X, Y, cv=5).mean())
print(cross_val_score(rfc, X, Y, cv=5).mean())
```
