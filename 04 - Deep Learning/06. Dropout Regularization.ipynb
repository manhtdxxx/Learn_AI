```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow as tf
from tensorflow import keras
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("sonar.csv", header=None)
df
```

```python
df.describe()
```

```python
df[60].value_counts()  # to check if binary or multi-class
```

```python
X = df.drop(60, axis='columns').values
Y = df[60].values
X.shape, Y.shape
```

```python
Y_dummies = pd.get_dummies(Y, drop_first=True, dtype='int16')
print(Y_dummies.shape)
Y_dummies = Y_dummies.values.reshape(-1)
Y_dummies.shape
```

```python
Y_dummies[:5], Y[:5]  # 1 is R, 0 is M
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y_dummies, train_size=0.8, random_state=5)
```

```python
values, counts = np.unique(y_train, return_counts=True)  # to ensure that we train a balanced training set
values, counts
```

```python
y_train.shape, y_test.shape
```

# Without Dropout

```python
model_1 = keras.Sequential([
    keras.layers.Input(shape=(60, )),             
    keras.layers.Dense(45, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model_1.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_1.fit(X_train, y_train, epochs=100)
```

```python
model_1.evaluate(X_test, y_test)  # training set has accuracy: 0.96
```

### accuracy of testing set is too low ==> OVERFITTING

---

# Using Dropout Regulalization

```python
model_2 = keras.Sequential([
    keras.layers.Input(shape=(60, )),             
    keras.layers.Dense(45, activation='relu', kernel_regularizer=keras.regularizers.L2(0.01)),  
    # 0.01 is the value of lambda, and this parameter applies a penalty on the weights of each neuron in this layer 
    # << including those (weights **2) * lambda in the cost function >>
    keras.layers.Dropout(0.2),  
    # Drop 20% of neurons in hidden layer 1
    keras.layers.Dense(30, activation='relu', kernel_regularizer=keras.regularizers.L2(0.01)),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation='sigmoid'),
])

model_2.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model_2.fit(X_train, y_train, epochs=100)
```

```python
model_2.evaluate(X_test, y_test)   # traning set has accuracy: 0.8311
```

### accuracy of testing set is better ==> LESS / NO OVERFITTING
