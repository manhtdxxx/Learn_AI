```python
from math import e
import matplotlib.pyplot as plt
import numpy as np
```

### Sigmoid / Log Function (0 <=  y  <= 1)

```python
def sigmoid(x):
    return 1 / (1 + e**-x)
```

```python
x1 = np.linspace(-10,10,100)
y1 = sigmoid(x1)

plt.plot(x1,y1)
plt.title('Sigmoid Function')
plt.grid(True)
```

### Tanh Function (-1 <= y <= 1)

```python
def tanh(x):
    return (e**x - e**(-x)) / (e**x + e**(-x))
```

```python
x2 = np.linspace(-10,10,100)
y2 = tanh(x2)

plt.plot(x2,y2)
plt.title('Tanh Function')
plt.grid(True)
```

### ReLU

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x3 = np.linspace(-10,10,100)
y3 = relu(x3)

plt.plot(x3,y3)
plt.title('ReLU Function')
plt.grid(True)
```

### Leaky ReLU

```python
def relu(x):
    return np.maximum(0.1*x,x)
```

```python
x4 = np.linspace(-100,40,100)
y4 = relu(x4)

plt.plot(x4,y4)
plt.title('Leaky ReLU Function')
plt.grid(True)
```
